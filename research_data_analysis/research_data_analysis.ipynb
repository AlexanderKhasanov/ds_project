````python
def get_size_flat(total_area):
    '''
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–≤–∞—Ä—Ç–∏—Ä –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
    '''
    if total_area < 30:
        return 'small'
    elif total_area < 70:
        return 'medium'
    elif total_area < 100:
        return 'big'
    return 'huge'
````

````python
# –†–∞—Å–ø—Ä–µ–¥–µ–ª–∏–º –∫–≤–∞—Ä—Ç–∏—Ä—ã –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
data['size_flat'] = data['total_area'].apply(get_size_flat)
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏ –æ—Ç –æ–±—â–µ–π
data['percent_living_area'] = data['living_area'] / data['total_area']
````

````python
for name, group_data in data.groupby('size_flat'):
    hist_name = f'kitchen_area - {name}'
    group_data.plot(y='kitchen_area', kind='hist', title=hist_name)
````

````python
for name, group_data in data.groupby('size_flat'):
    hist_name = f'percent_living_area - {name}'
    group_data.plot(y='percent_living_area', kind='hist', title=hist_name)
````

–î–∞–ª–µ–µ –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –æ–ø—Ä–µ–¥–µ–ª–∏–º –º–µ–¥–∏–∞–Ω—É –ø—Ä–æ—Ü–µ–Ω—Ç–∞ ```living_area``` –æ—Ç ```total_area```. –¢–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–∏–º –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ ```kitchen_area``` –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–æ–ø—É—Å–∫–æ–≤.
–ú–µ–¥–∏–∞–Ω–∞ –±—ã–ª–∞ –≤—ã–±—Ä–∞–Ω–∞ –≤ —Å–≤—è–∑–∏ —Å —Ç–µ–º, —á—Ç–æ –Ω–∞ –±–æ–ª—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤—ã–±—Ä–æ—Å—ã, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–≥—É—Ç —Å–∏–ª—å–Ω–æ –ø–æ–≤–ª–∏—è—Ç—å –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç

````python
for name, group_data in data.groupby('size_flat'):
    hist_name = f'kitchen_area - {name}'
    group_data.plot(y='kitchen_area', kind='hist', title=hist_name)
````

````python
# –ó–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ living_area
for category in data['size_flat'].unique():
    data.loc[(data['size_flat'] == category) & (data['living_area'].isna()), 'living_area'] = \
    data.loc[data['size_flat'] == category, 'living_area'].median()
# –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ –±–æ–ª—å—à–µ –Ω–µ—Ç –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–µ
data['living_area'].isna().sum()
````

````python
# –ó–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ kitchen_area
for category in data['size_flat'].unique():
    data.loc[(data['size_flat'] == category) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[data['size_flat'] == category, 'kitchen_area'].median()
# –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ –±–æ–ª—å—à–µ –Ω–µ—Ç –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–µ
data['living_area'].isna().sum()
````
